```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read data and add length columns

dataDir = '../'
csvFile = 'protein_seqs_cleaned_6types_reps_reserve.csv'

dataDf = pd.read_csv( os.path.join( dataDir, csvFile ) )
dataDf['siteLen'] = dataDf['site'].map( lambda x: len(x) )
dataDf['seqLen'] = dataDf['sequence'].map( lambda x: len(x) )

print(dataDf)
print(dataDf.describe())
dataDf.hist(bins=20)
print('unique sequences:', len(set(dataDf.sequence)) )
print('unique sites:', len(set(dataDf.site)) )
```

```python
# Assuming your dataframe is called 'df'
# df = pd.read_csv('your_data.csv')  # Load your data if needed

def analyze_enzyme_types(df):
    """
    Analyze enzyme type distribution in the dataframe
    
    Parameters:
    df : pandas.DataFrame
        DataFrame with columns: 'name', 'type', 'site', 'sequence'
    
    Returns:
    unique_types : list
        List of unique enzyme types
    type_counts : pandas.Series
        Count of each enzyme type
    """
    
    # 1. Get list of unique enzyme types
    unique_types = df['type'].unique().tolist()
    print("Unique enzyme types found:")
    print("-" * 30)
    for i, enzyme_type in enumerate(unique_types, 1):
        print(f"{i}. {enzyme_type}")
    print(f"\nTotal number of unique enzyme types: {len(unique_types)}")
    
    # Get counts for each type
    type_counts = df['type'].value_counts()
    print("\nEnzyme type distribution:")
    print("-" * 30)
    print(type_counts)
    
    # 2. Create histogram showing number of entries of each enzyme type
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Create bar plot
    type_counts.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
    
    # Customize the plot
    ax.set_title('Distribution of DNA Binding Enzyme Types', fontsize=16, fontweight='bold')
    ax.set_xlabel('Enzyme Type', fontsize=12)
    ax.set_ylabel('Number of Entries', fontsize=12)
    ax.grid(axis='y', alpha=0.3, linestyle='--')
    
    # Rotate x-axis labels if needed for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Add value labels on top of each bar
    for i, (idx, val) in enumerate(type_counts.items()):
        ax.text(i, val + 0.5, str(val), ha='center', va='bottom', fontsize=10)
    
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    
    # Show the plot
    plt.show()
    
    return unique_types, type_counts


# Alternative: If you want a horizontal bar chart for better label readability
def analyze_enzyme_types_horizontal(df):
    """
    Alternative visualization with horizontal bars (useful for long enzyme type names)
    """
    
    # Get unique types and counts
    unique_types = df['type'].unique().tolist()
    type_counts = df['type'].value_counts()
    
    # Create horizontal bar plot
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Sort by count for better visualization
    type_counts_sorted = type_counts.sort_values()
    
    # Create horizontal bar plot
    type_counts_sorted.plot(kind='barh', ax=ax, color='teal', edgecolor='black')
    
    # Customize the plot
    ax.set_title('Distribution of DNA Binding Enzyme Types', fontsize=16, fontweight='bold')
    ax.set_xlabel('Number of Entries', fontsize=12)
    ax.set_ylabel('Enzyme Type', fontsize=12)
    ax.grid(axis='x', alpha=0.3, linestyle='--')
    
    # Add value labels at the end of each bar
    for i, (idx, val) in enumerate(type_counts_sorted.items()):
        ax.text(val + 0.5, i, str(val), ha='left', va='center', fontsize=10)
    
    plt.tight_layout()
    plt.show()
    
    return unique_types, type_counts
```

```python
# Example usage with sample data (remove this when using your actual data)


# Run the analysis
print("Running analysis on sample data...")
print("=" * 50)
unique_types, type_counts = analyze_enzyme_types(dataDf)

# Uncomment below to see horizontal version
# analyze_enzyme_types_horizontal(df_sample)
```

```python

def plot_length_distributions_by_type(df):
    """
    Plot histograms of site lengths and sequence lengths for each enzyme type
    
    Parameters:
    df : pandas.DataFrame
        DataFrame with columns: 'name', 'type', 'site', 'sequence', 'siteLen', 'seqLen'
    """
    
    # Get unique enzyme types
    unique_types = df['type'].unique()
    n_types = len(unique_types)
    
    print(f"\nGenerating length distribution histograms for {n_types} enzyme types...")
    print("=" * 60)
    
    # Process each enzyme type
    for enzyme_type in unique_types:
        # Filter dataframe for current enzyme type
        df_subset = df[df['type'] == enzyme_type]
        n_entries = len(df_subset)
        
        print(f"\nEnzyme Type: {enzyme_type}")
        print(f"Number of entries: {n_entries}")
        print(f"Site length range: {df_subset['siteLen'].min()}-{df_subset['siteLen'].max()}")
        print(f"Sequence length range: {df_subset['seqLen'].min()}-{df_subset['seqLen'].max()}")
        
        # Create figure with two subplots side by side
        fig, axes = plt.subplots(1, 2, figsize=(12, 4))
        
        # Use pandas hist method to plot histograms
        # Plot site lengths
        df_subset[['siteLen']].hist(ax=axes[0], bins=20, color='coral', 
                                     edgecolor='black', alpha=0.7)
        axes[0].set_title(f'Site Length Distribution\n{enzyme_type}', fontsize=12, fontweight='bold')
        axes[0].set_xlabel('Site Length (bp)', fontsize=10)
        axes[0].set_ylabel('Frequency', fontsize=10)
        axes[0].grid(axis='y', alpha=0.3, linestyle='--')
        
        # Add statistics text
        mean_site = df_subset['siteLen'].mean()
        std_site = df_subset['siteLen'].std()
        axes[0].text(0.65, 0.95, f'Mean: {mean_site:.1f}\nStd: {std_site:.1f}\nN: {n_entries}',
                    transform=axes[0].transAxes, fontsize=9,
                    verticalalignment='top',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        
        # Plot sequence lengths
        df_subset[['seqLen']].hist(ax=axes[1], bins=20, color='skyblue', 
                                    edgecolor='black', alpha=0.7)
        axes[1].set_title(f'Protein Sequence Length Distribution\n{enzyme_type}', 
                         fontsize=12, fontweight='bold')
        axes[1].set_xlabel('Sequence Length (aa)', fontsize=10)
        axes[1].set_ylabel('Frequency', fontsize=10)
        axes[1].grid(axis='y', alpha=0.3, linestyle='--')
        
        # Add statistics text
        mean_seq = df_subset['seqLen'].mean()
        std_seq = df_subset['seqLen'].std()
        axes[1].text(0.65, 0.95, f'Mean: {mean_seq:.1f}\nStd: {std_seq:.1f}\nN: {n_entries}',
                    transform=axes[1].transAxes, fontsize=9,
                    verticalalignment='top',
                    bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
        
        # Add main title for the figure
        fig.suptitle(f'Length Distributions for {enzyme_type} (n={n_entries})', 
                    fontsize=14, fontweight='bold', y=1.02)
        
        # Adjust layout
        plt.tight_layout()
        plt.show()


def plot_length_distributions_grid(df, cols=4):
    """
    Alternative: Plot all enzyme types in a grid layout for comparison
    
    Parameters:
    df : pandas.DataFrame
        DataFrame with columns: 'name', 'type', 'site', 'sequence', 'siteLen', 'seqLen'
    cols : int
        Number of columns in the grid
    """
    
    unique_types = df['type'].unique()
    n_types = len(unique_types)
    rows = (n_types + cols - 1) // cols  # Calculate rows needed
    
    # Create figure with subplots for site lengths
    fig1, axes1 = plt.subplots(rows, cols, figsize=(cols*3, rows*3))
    fig1.suptitle('Site Length Distributions by Enzyme Type', fontsize=16, fontweight='bold')
    axes1 = axes1.flatten() if n_types > 1 else [axes1]
    
    # Create figure with subplots for sequence lengths
    fig2, axes2 = plt.subplots(rows, cols, figsize=(cols*3, rows*3))
    fig2.suptitle('Protein Sequence Length Distributions by Enzyme Type', fontsize=16, fontweight='bold')
    axes2 = axes2.flatten() if n_types > 1 else [axes2]
    
    for i, enzyme_type in enumerate(unique_types):
        # Filter dataframe for current enzyme type
        df_subset = df[df['type'] == enzyme_type]
        
        # Plot site lengths
        df_subset[['siteLen']].hist(ax=axes1[i], bins=15, color='coral', 
                                     edgecolor='black', alpha=0.7)
        axes1[i].set_title(f'{enzyme_type}\n(n={len(df_subset)})', fontsize=10)
        axes1[i].set_xlabel('Site Length', fontsize=8)
        axes1[i].set_ylabel('Frequency', fontsize=8)
        axes1[i].tick_params(labelsize=7)
        
        # Plot sequence lengths
        df_subset[['seqLen']].hist(ax=axes2[i], bins=15, color='skyblue', 
                                    edgecolor='black', alpha=0.7)
        axes2[i].set_title(f'{enzyme_type}\n(n={len(df_subset)})', fontsize=10)
        axes2[i].set_xlabel('Sequence Length', fontsize=8)
        axes2[i].set_ylabel('Frequency', fontsize=8)
        axes2[i].tick_params(labelsize=7)
    
    # Hide unused subplots
    for j in range(i+1, len(axes1)):
        axes1[j].axis('off')
        axes2[j].axis('off')
    
    plt.figure(fig1.number)
    plt.tight_layout()
    plt.show()
    
    plt.figure(fig2.number)
    plt.tight_layout()
    plt.show()


print(f"Total entries: {len(dataDf)}")
print(f"Columns: {list(dataDf.columns)}")
print("\n")

# Run the analysis
print("Step 1: Analyzing enzyme type distribution")
print("=" * 60)
unique_types, type_counts = analyze_enzyme_types(dataDf)

print("\n\nStep 2: Plotting length distributions for each enzyme type")
print("=" * 60)
plot_length_distributions_by_type(dataDf)

# Uncomment below to see grid layout version
# print("\n\nAlternative: Grid layout for comparison")
# print("=" * 60)
# plot_length_distributions_grid(df_sample, cols=4)
```

```python

# Uncomment below to see grid layout version
print("\n\nAlternative: Grid layout for comparison")
print("=" * 60)
plot_length_distributions_grid(dataDf, cols=4)
```

```python
# list of types to remove

removeTypes=[ 'orphan methyltransferase',
 'Type II methyl-directed restriction enzyme',
 'homing endonuclease',
 'control protein',
 'Type II specificity subunit',
 'Type I restriction enzyme',
 'Type III restriction enzyme',
 'Type II nicking endonuclease',
 'Type IV methyl-directed restriction enzyme',
 'Type I restriction enzyme/methyltransferase']


temp = dataDf.copy()
for etype in removeTypes:
    temp = temp[ temp.type != etype ]

temp.type.unique().tolist()
```

```python
# save it here
# output file names --- 'None' if not saving reformated data
# fix the naming issue in the dataframe first, then save

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

combinedFileOutput = 'protein_seqs_cleaned_6types.csv'
fastaFileName = 'protein_seqs_cleaned_6types.fasta'

#########################################################################################
if combinedFileOutput:
    temp[['RE', 'type', 'site','sequence' ]].to_csv(combinedFileOutput,index=False)

if fastaFileName:     # ---- Convert to SeqRecord objects ----
    records = []
    i=0
    for _, row in temp.iterrows():
        seq = Seq(str(row["sequence"]).strip())

        # check to make sure no double names!
        if row.RE.count('(')>0:
            print('ERROR! a name with (')
            print(row)
            
        records.append(
                    SeqRecord(seq, id=str(row.RE), description=row.site+' '+row.type)
                    )
        
    # ---- Write to FASTA ----
    SeqIO.write(records, fastaFileName, "fasta")
```
