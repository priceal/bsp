```python
# DEFINE Function to calculate edit distance between two sequences
# USE THE NORMALIZED LEVENSHTEIN DISTANCE
# (claude opus 4.1 output)

def sequence_distance(seq1, seq2):
    """
    Calculate normalized edit distance between two sequences.
    Uses Levenshtein distance normalized by the length of the longer sequence.
    """
    len1, len2 = len(seq1), len(seq2)
    
    # Create distance matrix
    d = np.zeros((len1 + 1, len2 + 1))
    
    # Initialize first column and row
    for i in range(len1 + 1):
        d[i][0] = i
    for j in range(len2 + 1):
        d[0][j] = j
    
    # Fill in the rest of the matrix
    for i in range(1, len1 + 1):
        for j in range(1, len2 + 1):
            if seq1[i-1] == seq2[j-1]:
                cost = 0
            else:
                cost = 1
            d[i][j] = min(
                d[i-1][j] + 1,      # deletion
                d[i][j-1] + 1,      # insertion
                d[i-1][j-1] + cost  # substitution
            )
    
    # Normalize by the length of the longer sequence
    return d[len1][len2] / max(len1, len2)

# manual addition for a reverse-complement symmetric distance metric
def sequenceDistanceRC( seq1, seq2 ):
    forward = sequence_distance( seq1, seq2 )
    reverseComp = sequence_distance( seq1, reverse_complement(seq2) )
    return min(forward,reverseComp)
    
```

```python
s1 = 'ACGTNX'
s2 = 'AXGXTXCXN'
rs1 = reverse_complement(s1)
rs2 = reverse_complement(s2)

print('sequence distance')
print(f'{s1} to {s2}:', sequence_distance(s1,s2))
print(f'{s2} to {s1}:', sequence_distance(s2,s1))
print(f'{s1} to {rs2}:', sequence_distance(s1,rs2))
print(f'{rs2} to {s1}:', sequence_distance(rs2,s1))
print(f'{rs1} to {s2}:', sequence_distance(rs1,s2))
print(f'{s2} to {rs1}:', sequence_distance(s2,rs1))
print(f'{rs1} to {rs2}:', sequence_distance(rs1,rs2))
print(f'{rs2} to {rs1}:', sequence_distance(rs2,rs1))

print('reverse-complement symmetric distance')
print(f'{s1} to {s2}:', sequenceDistanceRC(s1,s2))
print(f'{s2} to {s1}:', sequenceDistanceRC(s2,s1))
print(f'{s1} to {rs2}:', sequenceDistanceRC(s1,rs2))
print(f'{rs2} to {s1}:', sequenceDistanceRC(rs2,s1))
print(f'{rs1} to {s2}:', sequenceDistanceRC(rs1,s2))
print(f'{s2} to {rs1}:', sequenceDistanceRC(s2,rs1))
print(f'{rs1} to {rs2}:', sequenceDistanceRC(rs1,rs2))
print(f'{rs2} to {rs1}:', sequenceDistanceRC(rs2,rs1))
```

```python
# Create distance matrix for all unique sequences
n_unique = len(unique_sites)
distance_matrix = np.zeros((n_unique, n_unique))

print("Calculating pairwise distances...")
for i in range(n_unique):
    if i % 100 == 0:
        print(f"Progress: {i}/{n_unique}")
    for j in range(i+1, n_unique):
        dist = sequenceDistanceRC(unique_sites[i], unique_sites[j])
        distance_matrix[i][j] = dist
        distance_matrix[j][i] = dist
print("...done")
```
